```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

```python
def LotkaVolterrra(y, t, parameters):
    return [parameters[0]*y[0] - parameters[1]*y[0]*y[1],
            parameters[3]*y[0]*y[1] - parameters[2]*y[1]]
```

```python
p = [0.6, 0.1, 1.5, 0.75]
y0 = [1.0, 1.0]
t = np.linspace(0, 20, 1000)
```

```python
res = odeint(LotkaVolterrra, y0, t, args=(p,))
```

```python
plt.plot(t, res[:,0], label = "rabbits")
plt.plot(t, res[:,1], label = "foxes")
plt.legend()
plt.show()
```
